```python
from IPython.display import clear_output
from datetime import datetime, timedelta

def limpar_tela():
    clear_output()

def main():
    limpar_tela()
    print("╔══════════════════════════════╗")
    print("║   APÓLICE DE SEGURO DE VIDA  ║")
    print("╚══════════════════════════════╝\n")

    # Dados Pessoais
    print("=== DADOS DO SEGURADO ===")
    nome = input("Nome completo: ").title()

    # Validação de Idade
    while True:
        try:
            idade = int(input("Idade: "))
            if 0 <= idade <= 120:
                break
            print("Idade inválida! Digite entre 0 e 120")
        except ValueError:
            print("Digite apenas números!")

    # Validação de CPF
    while True:
        cpf = input("CPF (somente números): ")
        cpf_limpo = ''.join(filter(str.isdigit, cpf))
        if len(cpf_limpo) == 11:
            cpf_formatado = f"{cpf_limpo[:3]}.{cpf_limpo[3:6]}.{cpf_limpo[6:9]}-{cpf_limpo[9:]}"
            break
        print("CPF inválido! Deve conter 11 dígitos.")

    # Validação de RG
    while True:
        rg = input("RG (somente números): ")
        rg_limpo = ''.join(filter(str.isdigit, rg))
        if len(rg_limpo) == 9:
            rg_formatado = f"{rg_limpo[:2]}.{rg_limpo[2:5]}.{rg_limpo[5:8]}-{rg_limpo[8:]}"
            break
        print("RG inválido! Deve conter 9 dígitos.")

    nome_mae = input("Nome da mãe: ").title()

    # Questionário Médico
    limpar_tela()
    print("=== QUESTIONÁRIO MÉDICO ===")
    fumante = input("Fumante? (S/N): ").upper() == 'S'
    doenca = input("Doença crônica? (S/N): ").upper() == 'S'

    # Estado Civil
    print("\nESTADO CIVIL:")
    print("1 - Casado(a)")
    print("2 - Solteiro(a)")
    print("3 - Viúvo(a)")
    print("4 - Divorciado(a)")
    print("5 - União Estável")

    while True:
        opcao = input("Opção (1-5): ")
        if opcao in ('1', '2', '3', '4', '5'):
            estados = ['Casado(a)', 'Solteiro(a)', 'Viúvo(a)', 'Divorciado(a)', 'União Estável']
            estado_civil = estados[int(opcao)-1]
            break
        print("Opção inválida!")

    # Beneficiários
    limpar_tela()
    beneficiarios = []
    print("=== BENEFICIÁRIOS ===")

    while True:
        try:
            num_benef = int(input("Quantos beneficiários (0-5)? "))
            if 0 <= num_benef <= 5:
                break
            print("Digite entre 0 e 5!")
        except ValueError:
            print("Número inválido!")

    for i in range(num_benef):
        print(f"\nBeneficiário {i+1}:")
        nome_benef = input("Nome: ").title()
        parentesco = input("Parentesco: ").title()
        beneficiarios.append((nome_benef, parentesco))

    # Cálculo do Prêmio
    valor_base = 30.00
    ajustes = []

    if idade < 25:
        valor_base *= 1.15
        ajustes.append("+15% (idade < 25)")
    elif idade > 60:
        valor_base *= 1.35
        ajustes.append("+35% (idade > 60)")

    if fumante:
        valor_base *= 1.25
        ajustes.append("+25% (fumante)")

    if doenca:
        valor_base *= 1.30
        ajustes.append("+30% (doença crônica)")

    if estado_civil == 'Casado(a)':
        valor_base *= 0.90
        ajustes.append("-10% (casado)")

    if estado_civil == 'União Estável':
        valor_base *= 0.95
        ajustes.append("-5% (união estável)")

    # Vigência
    hoje = datetime.now()
    inicio_vigencia = hoje + timedelta(days=7)
    fim_vigencia = inicio_vigencia + timedelta(days=365)

    # Exibição da Apólice
    limpar_tela()
    print("\n" + "="*40)
    print(f"{'APÓLICE DE SEGURO':^40}")
    print("="*40)
    print(f"\n{'SEGURADO':<15}: {nome}")
    print(f"{'IDADE':<15}: {idade} anos")
    print(f"{'CPF':<15}: {cpf_formatado}")
    print(f"{'RG':<15}: {rg_formatado}")
    print(f"{'MÃE':<15}: {nome_mae}")
    print("\nPERFIL:")
    print(f"{'FUMANTE':<15}: {'Sim' if fumante else 'Não'}")
    print(f"{'DOENÇA':<15}: {'Sim' if doenca else 'Não'}")
    print(f"{'ESTADO CIVIL':<15}: {estado_civil}")
    print("\nBENEFICIÁRIOS:")
    for i, (nome, parentesco) in enumerate(beneficiarios, 1):
        print(f"{i}. {nome} ({parentesco})")
    print("\nAJUSTES APLICADOS:")
    for ajuste in ajustes or ["Nenhum"]:
        print(f"- {ajuste}")
    print("\nVIGÊNCIA:")
    print(f"Início: {inicio_vigencia.strftime('%d/%m/%Y')}")
    print(f"Término: {fim_vigencia.strftime('%d/%m/%Y')}")
    print("\n" + "="*40)
    print(f"{'PRÊMIO MENSAL':<20}: R$ {valor_base:.2f}")
    print("="*40 + "\n")

if __name__ == "__main__":
    main()
```
